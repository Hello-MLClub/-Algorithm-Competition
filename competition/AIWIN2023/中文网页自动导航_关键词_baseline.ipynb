```python
import json
import pandas as pd
```

### 数据说明

数据分为训练数据、A榜数据、B榜数据（每类数据都包含指令instruction、标签label）：
- instruction_trainset.json ：训练集指令（1）
- label_trainset.json       ：训练集标签（2）
- instruction_testA.json    ：A榜指令（3）


```python
LABEL_PATH = './2023S-T1-A-Data[供选手]/指令数据&标签/'
```

```python
label_trainset = json.load(open(LABEL_PATH + 'label_trainset.json'))
instruction_trainset = json.load(open(LABEL_PATH + 'instruction_trainset.json'))
instruction_testA = json.load(open(LABEL_PATH + 'instruction_testA.json'))
```

```python
from collections import defaultdict, Counter
```

```python
submit_prediction = []
for idx in range(len(label_trainset)):
    golden_key_value = {}
    golden_start_tag_freq = defaultdict(int)
    page_source = instruction_trainset[idx]['page_source']
    
    for query, label in zip(
        instruction_trainset[idx]['instruction_detail'],
        label_trainset[idx]['instruction_detail']
    ):    
        for tag in label['key-value'].keys():
            if tag not in golden_key_value:
                golden_key_value[tag] = label['key-value'][tag]
                
            golden_start_tag_freq[tag] += 1
    
    golden_start_tag = Counter(golden_start_tag_freq).most_common(1)[0][0]
    
    page_predictions = []
    for instruction_detail in instruction_testA[idx]['instruction_detail']:
        prediction = {}
        prediction['instruction'] = instruction_detail['instruction']
        prediction['key-value'] = {}
        prediction['key-value'][golden_start_tag] = golden_key_value[golden_start_tag]
        
        for key in instruction_detail['key-value']:
            if key in golden_key_value:
                prediction['key-value'][key] = golden_key_value[key]
                
        page_predictions.append(prediction)
    
    submit_prediction.append({
        'page_source': page_source,
        'instruction_detail': page_predictions
    })
```

```python
with open('T2.json', 'w') as up:
    json.dump(submit_prediction, up)
```
